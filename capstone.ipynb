## Temperatura no Rio de Janeiro

O arquivo `data/23.31S-42.82W-TAVG-Trend.txt` contém dados de temperatura média mensal para a cidade do Rio de Janeiro. O arquivo também contém médias móveis anual, 5, 10 e 20 anos. Esses dados foram baixados do site Berkeley Earth (http://berkeleyearth.lbl.gov/locations/23.31S-42.82W).

### Tarefa

Faça duas funções, uma que lê os dados de temperatura mensal, outra que lê os dados da média móvel anual.
As duas funções devem:

* Receber como entrada **somente** o nome do arquivo de dados.
* Retornar duas listas: uma com as datas referentes aos dados e outra com os dados de temperatura.
* As datas retornadas devem ser em anos decimais. Ex: Janeiro de 1984 seria 1984.0833333333333 (1984 + 1/12).
* Datas sem valores de temperatura (NaN) devem ser ignoradas (não incluidas nas listas).

Utilize suas funções para carregar os dados e fazer um gráfico da temperatura média mensal e média movel anual pelo tempo.

```python
def temperatura_mensal_data(nomedoarquivodedados): 
    arquivo = open(nomedoarquivodedados) # comando para abrir o arquivo desejado
    data_mensal = [] 
    temp_mensal = []

    for linha in arquivo:
        lista = linha.split() 
        if linha[0] != "%": 

            if len(lista) != 0: 
                if lista[2] != "NaN": 

                    ano = float(lista[0]) 

                    mes = float(lista[1]) 

                    anos_decimais = ano + mes/12 
                    data_mensal.append(anos_decimais) 
                    anomalia_mensal = float(lista[2]) 
                    media_mensal = 24.01 + anomalia_mensal 
                    temp_mensal.append(media_mensal) 
    return data_mensal, temp_mensal 
```

```python
temperatura_mensal_data("data/23.31S-42.82W-TAVG-Trend.txt")
```

```python
def temperatura_anual_data(nomedoarquivodedados):
    arquivo = open(nomedoarquivodedados) 
    data_anual = [] 
    temp_anual = [] 
    for linha in arquivo: 
        lista = linha.split()
        if linha[0] != "%": 
            if len(lista) != 0:  
                if lista[4] != "NaN": 
                    ano = float(lista[0]) 
                    mes = float(lista[1])
                    anos_decimais = ano + mes/12
                    data_anual.append(anos_decimais) 
                    anomalia_anual = float(lista[4]) 
                    media_anual = 24.01 + anomalia_anual 
                    temp_anual.append(media_anual) 
    return data_anual, temp_anual
```

```python
temperatura_anual_data("data/23.31S-42.82W-TAVG-Trend.txt")
```

```python
[data_mensal, temp_mensal] = temperatura_mensal_data("data/23.31S-42.82W-TAVG-Trend.txt")
[data_anual, temp_anual] = temperatura_anual_data("data/23.31S-42.82W-TAVG-Trend.txt")
plt.figure() #comando para criar figura
plt.plot(data_mensal, temp_mensal, ".k", label = "Média mensal") 
plt.plot(data_anual, temp_anual, "-r", label = "Média móvel anual", linewidth = 2) 
plt.xlabel("Ano") 
plt.xlim(1832, 2014) 
plt.ylabel("Temperatura Média (°C)")
legend = plt.legend(loc = 'upper left', fontsize = 'large')
```

### Tarefa

Faça uma função que calcule a temperatura média anual a partir das temperaturas mensais. A sua função deve:

* Receber como entrada a lista das datas e a lista das temperaturas mensais.
* Retornar duas listas: uma com os anos e outra com as temperaturas médias correspondetes.
* Anos que não contem dados de todos os 12 meses devem ser ignorados (não incluídos nas listas retornadas).

Utilize sua função para calcular a média anual. Faça um gráfico da temperatura média anual por ano junto com a média móvel anual.

**Dica**: A função `math.floor` retorna o número inteiro que precede um número real. Ex: `math.floor(1984.23) == 1984`

```python
def temperatura_media_anual(datas, temperaturas_mensais):
    
    tempoanos = [] 
    temp_media_anual = [] 
    for i in range(0, len(datas) + 1):
        temp_anual_mensal = 0 
        if i + 11 < len(datas) and math.floor(datas[i]) == math.floor(datas[i + 11]): 
            for j in range(0, 12):
                temp_anual_mensal = temp_anual_mensal + (temperaturas_mensais[i + j]/12) 
            tempoanos.append(math.floor(datas[i])) 
            temp_media_anual.append(temp_anual_mensal)
    return tempoanos, temp_media_anual 
```

```python
temperatura_media_anual(data_mensal, temp_mensal)
```

```python
[anos, temp_media_anual] = temperatura_media_anual(data_mensal, temp_mensal) 
plt.figure()
plt.plot(anos, temp_media_anual, "ok", label = "Média anual") 
plt.plot(data_anual, temp_anual, "-r", label = "Média móvel anual")
plt.xlabel("Ano") 
plt.xlim(1832, 2014) 
plt.ylabel("Temperatura Média (°C)") 
legend = plt.legend(loc = 'upper left', fontsize = 'large')
```
